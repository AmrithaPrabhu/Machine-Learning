```python
import numpy as np
import pandas as pd
```

```python
class SVM():

  def __init__(self,learningRate,no_of_iterations,lambda_reg):
      self.learningRate=learningRate
      self.no_of_iterations=no_of_iterations
      self.lambda_reg=lambda_reg

  def fit(self,X,Y):
    self.m,self.n=X.shape
    self.w=np.zeros(self.n)
    self.b=0
    self.X=X
    self.Y=Y

    for i in range(self.no_of_iterations):
         self.update_weights()
  def update_weights(self):

      #label encoding
      y_label=np.where(self.Y<=0 , -1,1)
      for index, x_i in enumerate(self.X):

              condition = y_label[index] * (np.dot(x_i, self.w) - self.b) >= 1

              if (condition == True):

                dw = 2 * self.lambda_reg * self.w
                db = 0

              else:

                dw = 2 * self.lambda_reg * self.w - np.dot(x_i, y_label[index])
                db = y_label[index]


              self.w = self.w - self.learningRate * dw

              self.b = self.b - self.learningRate * db

  def predict(self, X):

    output = np.dot(X, self.w) - self.b
    
    predicted_labels = np.sign(output) #wherever its less than -1, value will become -1 and more than 1 becomes 1

    y_hat = np.where(predicted_labels <= -1, 0, 1) #reverting -1,1 to 0,1

    return y_hat
```

```python
from sklearn.metrics import accuracy_score
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
```

```python
diabetes_data=pd.read_csv("/content/diabetes.csv")
diabetes_data.head()
```

```python
diabetes_data.shape
```

```python
diabetes_data.describe()
```

```python
input=diabetes_data.drop(columns="Outcome",axis=1)
outcome=diabetes_data["Outcome"]
input.head()
```

```python
scalar=StandardScaler()
scalar.fit(input)
standard_data=scalar.transform(input)
```

```python
print(standard_data)
```

```python
features=standard_data
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(features,outcome,test_size=0.2,random_state=2)
```

```python
model=SVM(0.001,1000,0.1)
```

```python
model.fit(X_train,Y_train)
```

```python
modelResults=model.predict(X_train)
accuracy=accuracy_score(Y_train,modelResults)
print(accuracy)
```

```python
modelResults=model.predict(X_test)
accuracy=accuracy_score(Y_test,modelResults)
print(accuracy)
```
