```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
from sklearn.preprocessing import StandardScaler
```

```python
credit=pd.read_csv("/content/creditcard.csv")
```

```python
credit.head()
```

```python
credit.info()
```

```python
credit.isnull().sum()
```

```python
credit['Class'].value_counts()
```

```python
#very unbalanced data as 99% belongs to only one class
legit=credit[credit['Class']==0]
fraud=credit[credit['Class']==1]
```

```python
print(legit)
```

Under Sampling

```python
legit_sample=legit.sample(n=492)
```

```python
legit_sample.shape
```

```python
data=pd.concat([fraud,legit_sample],axis=0)
```

```python
data.head()
```

```python
data.tail()
```

```python
data.shape
```

```python
data['Class'].value_counts()
```

```python
data.groupby("Class").mean()
```

```python
X=data.drop(columns='Class')
Y=data['Class']
print(X)
print(Y)
X_train, X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2,random_state=2)
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
model=LogisticRegression()
```

```python
model.fit(X_train,Y_train)
```

```python
train_prediction=model.predict(X_train)
trainAccuracy=accuracy_score(train_prediction,Y_train)
print(trainAccuracy)
```

```python
test_prediction=model.predict(X_test)
testAccuracy=accuracy_score(test_prediction,Y_test)
print(testAccuracy)
```
